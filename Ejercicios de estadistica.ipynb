```python
mu = 2.0
sigma_1 = 5.0
sigma_2 = 2.0
muestras_1 = np.random.normal(loc = mu, scale = sigma_1, size = 400)
muestras_2 = np.random.normal(loc = mu, scale = sigma_2, size = 400)
print(muestras_1, muestras_2)
```

```python
plt.figure(figsize = (10,8))
plt.hist(muestras_1, bins = 20, alpha = 0.5, label = 'Histrograma Muestra 1')
plt.hist(muestras_2, bins = 20, alpha = 0.5, label = 'Histrograma Muestra 2')
plt.legend()
plt.grid()
plt.show()
```

¿Cómo se modifica si el dado está cargado?. Simular un dado cargado (pueden cargar el número que quieran). Pista: choice

```python
#Crear un arreglo que asigne probabilidades a cada valor del dado
prob=np.array([1,1,1,1,1,6])
prob=prob/prob.sum()
prob
```

```python
#Genera la muestra
mdc=np.random.choice([1,2,3,4,5,6],p=prob,size=1000)
mdc
```

```python
np.mean(mdc)
```

```python
np.std(mdc)
```
